```python
import pandas as pd
dataframe=pd.read_csv("diabetes.csv")
```

```python
dataframe.head()
```

```python
x=dataframe.drop("Outcome",1)
y=dataframe.Outcome
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=8)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
```

```python
tree=tree.fit(X_train,Y_train)
predictedby=tree.predict(X_test)
```

```python
from sklearn import metrics
accuracy=metrics.accuracy_score(Y_test,predictedby)
print(accuracy)
```

```python
from sklearn.tree import export_graphviz
from io import StringIO
```

```python
dotdata = StringIO()
export_graphviz(tree,out_file=dotdata,filled=True,rounded=True,special_characters=True,class_names=["0","1"])
print(dotdata.getvalue())
import pydotplus 
graph = pydotplus.graph_from_dot_data(dotdata.getvalue())
graph.write_png("tree_for_diabetes.png")
from IPython.display import Image
Image(graph.create_png())
```
